```python
import numpy as np
import pandas as pd
import requests
import json
```

```python
#create list heroes you want to find
heroes_list=['Anti-Mag',
'Axe',
'Bane',
'Bloodseeker',
'Crystal Maide',
'Drow Range', 
'Earthshaker',
'Mirana',
'Razor',
'Tiny']
```

```python
heroes_list
```

```python
#Call link API table heroes dota and export to json file
endpoint_get = 'https://api.opendota.com/api/heroes'
r = requests.get(endpoint_get)
hero_dota = r.json()
```

```python
#json file dota heroes
hero_dota
```

```python
#create a loop to find the appropriate heroes name in heroes_list from the json file hero_dota
#then save the search results into the data_hero list
data_hero=[]
for i in heroes_list:
    for j in range (0,len(hero_dota)):
        if i in hero_dota[j]['localized_name']:
            data_hero.append(hero_dota[j])
```

```python
data_hero
```

```python
#create dataframe df_hero from data_hero
df_hero=pd.DataFrame(data_hero)
```

```python
df_hero
```

# Enrich data heroes by get dota heroes statistics data

```python
#Call link API table heroes dota statistics and export to json file
endpoint_get = 'https://api.opendota.com/api/heroStats'
r = requests.get(endpoint_get)
hero_stat = r.json()
```

```python
#json file dota heroes statistics
hero_stat
```

```python
#create a loop to find the appropriate heroes name in heroes_list from the json file hero_stat
#then save the search results into the stat_hero list
stat_hero=[]
for i in heroes_list:
    for j in range (0,len(hero_stat)):
        if i in hero_stat[j]['localized_name']:
            stat_hero.append(hero_stat[j])
```

```python
stat_hero
```

```python
#create dataframe df_stat from stat_hero
df_stat_hero=pd.DataFrame(stat_hero)
```
